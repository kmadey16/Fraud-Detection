# Winning model metrics

## Random Forest model

1. Train test split done using stratify parameter to keep proportion of fraud and nonfraud labeled data in train and test sets.
2. GridSearchCV used for hyperparameter tuning, with the winning model using a max depth of 10 and number of estimators at 100.


## Final model features

The existing features - 
1. step (timestamp of the hour of the month that the transaction occured)
2. type of transaction (dummies created)
3. amount - amount of the transaction in local currency.
4. oldbalanceOrg - initial balance before the transaction
5. newbalanceOrig - new balance after the transaction
6. oldbalanceDest - initial balance recipient before the transaction. 
7. newbalanceDest - new balance recipient after the transaction. 


Added Features - 
1. Hour - hour of day the transaction occured
2. dayofweek - day of week the transaction occured
3. dayofmonth - day of month the transaction occured
4. inpeaktime - 1 if transaction occured during peak hours for fraud, else 0.
5. coupled - 1 if transaction was a transfer that had a cash out immediately after, else 0



Target feature - 
1. isFraud - This is the transactions made by the fraudulent agents inside the simulation. In this specific dataset the fraudulent behavior of the agents aims to profit by taking control or customers accounts and try to empty the funds by transferring to another account and then cashing out of the system.


### Winning parameters and hyperparameters

As mentioned before, the winning model had hyperparameters for the random forest model as max depth of 10 and num estimators at 100.

The RF classifier had a balanced class weight and the GridSearchCV had a scoring parameter f1.
